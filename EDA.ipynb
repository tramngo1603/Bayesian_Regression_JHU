```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_link = "https://raw.githubusercontent.com/rashida048/Datasets/master/home_data.csv"

# import the dataset
df = pd.read_csv(dataset_link)
# shape of data
print(df.shape)
#first 5 rows of the dataset
df.head()
```

```python
# basic information about the dataset (numerical data)
df.describe()
```

```python
# drop id and date 
df = df.drop(["id", "date"], axis=1)
df.head()
```

```python
df.hist(bins = 50, figsize =(30,20))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
%matplotlib inline

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

#Create a copy of this training set to play with.
housing = train_set.copy()
```

```python
corr_matrix = housing.corr()
corr_matrix["price"].sort_values(ascending = False)
```

```python
attributes = ["price", "sqft_living", "grade", "sqft_above", "sqft_living15", "bathrooms", "view"]
scatter_matrix(housing[attributes], figsize = (12,8), alpha = 0.2);
```

```python
housing.plot(kind="scatter", x="sqft_living", y="price",alpha=0.2);
```

```python
housing.plot(kind="scatter", x="grade", y="price", alpha=0.2);
```
